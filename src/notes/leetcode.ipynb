```python
# Bubble Sort
import time
def bubble_sort(ls):
    """bubble sort.

    :param ls: list, to be sortted list.
    """
    for j in range(len(ls)-1):
        for i in range((len(ls)-j)-1):
            if ls[i] > ls[i+1]:
                ls[i], ls[i+1] = ls[i+1], ls[i]

unsort = [9,8,7,6,5,4,3,2,1,0]*10
start_time = time.process_time()
bubble_sort(unsort)
# print("耗时: %.6f 秒" % (time.process_time()-start_time), "\n结果: ", unsort)
print("耗时: %.6f 秒" % (time.process_time()-start_time))
```

```python
# Quick Sort
import time
import random

def quick_sort(ls):
    """quick Sort

    :param ls: list, to be sortted list.
    """
    if len(ls) < 2:
        return ls
    pivot_index = random.randrange(len(ls))
    pivot = ls[pivot_index]
    greater = []
    lesser = []

    for i in (ls[:pivot_index]+ls[pivot_index+1:]):
        lesser.append(i) if i <= pivot else greater.append(i)

    return quick_sort(lesser)+[pivot]+quick_sort(greater)

unsort = [13,24,12,5,6,4,8,1,7,35,23,26,10]*100
start_time = time.process_time()
sorted_ls = quick_sort(unsort)
print("耗时: %.6f 秒" % (time.process_time()-start_time))
```

```python
# Select Sort
import time
def select_sort(ls):
    for j in range(len(ls)):
        min_index = j
        for i in range(j, len(ls)):
            if ls[i] < ls[min_index]:
                min_index = i
        if min_index != j:
            ls[j], ls[min_index] = ls[min_index], ls[j]

unsort = [13,24,12,5,6,4,8,1,7,35,23,26,10]*100
start_time = time.process_time()
select_sort(unsort)
print("耗时: %.6f秒 \n" % (time.process_time()-start_time))
```

```python
# Inser Sort
import time
def insert_sort(ls):
    for i in range(1, len(ls)):
        insert_index = i
        insert_val = ls[i]
        while insert_index>=1 and insert_val < ls[insert_index-1]:
            ls[insert_index], ls[insert_index-1] = ls[insert_index-1], ls[insert_index]
            insert_index -= 1

unsort = [13,24,12,5,6,4,8,1,7,35,23,26,10]*100
start_time = time.process_time()
insert_sort(unsort)
print("耗时 %.6f 秒" % (time.process_time()-start_time))
```

```python
# Heap Sort
import time

def heapify(arr, n, i):
    largest = i  # 初始化设置最大数位置为i
    l = 2 * i + 1  # i的左子节点
    r = 2 * i + 2  # i的右子节点

    if l < n and arr[l] > arr[largest]:  # i与左子节点比较
        largest = l

    if r < n and arr[r] > arr[largest]:  # i与右子节点比较
        largest = r

    if largest != i:
        arr[largest], arr[i] = arr[i], arr[largest]  # 将i与最大子节点数值互换
        heapify(arr, n, largest)  # 交换数值后，递归处理该分支所有子树，确保子树都是最小堆

def heap_sort(arr):
    n = len(arr)

    for i in range(n//2 -1, -1, -1):  # 从最大非子节点位置遍历到根节点位置
        heapify(arr, n, i)  # 构建最大堆

    # 最大堆已经形成，根节点位置肯定为最大值，总堆开始排序
    for i in range(n-1, 0, -1):  # 从最后一个位置遍历到根节点位置
        arr[i], arr[0] = arr[0], arr[i]  # 将根节点（最大值）放到节点i位置
        heapify(arr, i, 0)  # 交换后再次构建最大堆

unsort = [13,24,12,5,6,4,8,1,7,35,23,26,10]*100
start_time = time.process_time()
heap_sort(unsort)
print("耗时: %.6f 秒" % (time.process_time()-start_time))
```

```python
# BiTree
import queue
import random


class TreeNode:
    def __init__(self, data):
        self.data = data
        self.right = None
        self.left = None


def build_tree(n):
    """Generate a binary tree at random

    :param n: int, size of the binary tree.
    :return: TreeNode, root node of the binary tree.
    """
    q: queue.Queue = queue.Queue()
    tree_node = TreeNode(random.randint(1, 20))
    q.put(tree_node)
    counter = n
    while not q.empty() and counter > 0:
        node_found = q.get()
        weight = [0, 1, 1, 1]
        if weight.pop(random.randrange(len(weight))):
            left_node = TreeNode(random.randint(1, 20))
            node_found.left = left_node
            counter -= 1
            q.put(left_node)
        if weight.pop(random.randrange(len(weight))):
            right_node = TreeNode(random.randint(1, 20))
            node_found.right = right_node
            counter -= 1
            q.put(right_node)


    return tree_node

node = build_tree(20)

# PreOrder-Recursive
def pre_order(node: TreeNode):
    if not node:
        return
    print(node.data, end=", ")
    pre_order(node.left)
    pre_order(node.right)

print("先序: ", end="")
pre_order(node)
print()

# InOrder-Recursive
def in_order(node: TreeNode):
    if not node:
        return
    in_order(node.left)
    print(node.data, end=", ")
    in_order(node.right)

print("中序: ", end="")
in_order(node)
print()

# PostOrder-Recursive
def post_order(node: TreeNode):
    if not node:
        return
    post_order(node.left)
    post_order(node.right)
    print(node.data, end=", ")

print("后序: ", end="")
post_order(node)
print()

# LevelOrder
def level_order(node: TreeNode):
    if not isinstance(node, TreeNode):
        return
    q = queue.Queue()
    q.put(node)
    while not q.empty():
        queued_node = q.get()
        print(queued_node.data, end=", ")
        if queued_node.left:
            q.put(queued_node.left)
        if queued_node.right:
            q.put(queued_node.right)

print("层序: ", end="")
level_order(node)
print()

# PreOrder Iter
def pre_order_iter(node: TreeNode):
    if not isinstance(node, TreeNode) or not node:
        return
    stack = [node]
    while stack:
        _node = stack.pop()
        print(_node.data, end=", ")
        if _node.right:
            stack.append(_node.right)
        if _node.left:
            stack.append(_node.left)
print("先序迭代: ", end="")
pre_order_iter(node)
print()

# InOrder Iter
def in_order_iter(node: TreeNode):
    if not isinstance(node, TreeNode) or not node:
        return
    stack = []
    n = node
    while n or stack:
        while n:
            stack.append(n)
            n = n.left
        n = stack.pop()
        print(n.data, end=", ")
        n = n.right

print("中序迭代: ", end="")
in_order_iter(node)
print()

# PostOrder Iter
def post_order_iter(node: TreeNode):
    res = []
    if not isinstance(node, TreeNode) or not node:
        return res
    stack = [node]
    while stack:
        _node = stack.pop()
        res.append(_node.data)
        if _node.left:
            stack.append(_node.left)
        if _node.right:
            stack.append(_node.right)
    return res[::-1]  # reversed

print("后序迭代: ", end="")
print(", ".join(map(str, post_order_iter(node))))
print()
```
